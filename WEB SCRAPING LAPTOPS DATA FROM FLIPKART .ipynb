# Web Scraping Flipkart Laptop Data

### Importing Required Libraries

```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
from requests import get
```

```python
HEADERS=({'User-Agent':'Chrome/93.0.4577.82','Accept-Language':'en-US en;q=0.5'})
```

### Website URL

```python
url='https://www.flipkart.com/search?q=laptop&sid=6bo%2Cb5g&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_3_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_3_na_na_na&as-pos=1&as-type=RECENT&suggestionId=laptop%7CLaptops&requestId=28a5d22e-0006-4564-9a76-cd4bbbe0bd45&as-backfill=on'
page=get(url,headers=HEADERS)
print(page.text)
```

```python
soup=BeautifulSoup(page.content,'html.parser')
```

```python
name=[]
randr=[]
price=[]
op=[]
dis=[]
```

```python
res=soup.find_all('div',class_='_4rR01T')
for r in res:
    name.append(r.text)
```

```python
len(name)
```

```python
res1=soup.find_all('span',class_='_2_R_DZ')
for r in res1:
    randr.append(r.text)
```

```python
len(randr)
```

```python
res2=soup.find_all('div',class_='_30jeq3 _1_WHN1')
for r in res2:
    price.append(r.text)
```

```python
len(price)
```

```python
res3=soup.find_all('div',class_='_3I9_wc _27UcVY')
for r in res3:
    op.append(r.text)
```

```python
len(op)
```

```python
op.append(0)
```

```python
len(op)
```

```python
res4=soup.find_all('div',class_='_3Ay6Sb')
for r in res4:
    dis.append(r.text)
```

```python
len(dis)
```

```python
dis.pop()
```

```python
dis.append(0)
```

```python
len(dis)
```

```python
df=pd.DataFrame({'Device Company and Name':name,'Ratings & Reviews':randr,'Current Price':price,
                'Original Price':op,'Discount':dis})
df
```

```python
df['Original Price']=np.where(df['Original Price']==0,df['Current Price'],df['Original Price'])
```

```python
df['Discount']=np.where(df['Discount']==0,'0% off',df['Discount'])
```

```python
df
```

### Spliting Ratings and Reviews Columns 

```python
new=df['Ratings & Reviews'].str.split('&',n=1,expand=True)
df['Ratings']=new[0]
df['Reviews']=new[1]
df.drop('Ratings & Reviews',axis=1,inplace=True)
```

### Removing White Spaces 

```python
df['Ratings']=df['Ratings'].str.rstrip()
df['Reviews']=df['Reviews'].str.lstrip()
```
